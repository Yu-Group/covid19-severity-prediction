## Example usage to pull a particular data source

```python
import os
from os.path import join as oj
```

```python
# data source of interest
data_source = "chrr_health"
```

```python
# current working directory
cur_dir = os.getcwd()
```

```python
# raw data directory
raw_dir = oj(cur_dir, "county_level", "raw", data_source)
```

```python
# processed data directory
processed_dir = oj(cur_dir, "county_level", "processed", data_source)
```

### 1. Download Data

```python
os.chdir(raw_dir)  # changing directories to download data to raw data source folder
import download
os.chdir(cur_dir)  # changing directories back to original folder
```

Raw data is downloaded in data/county_level/raw/\<data source\> folder

### 2. Load Data

```python
# loading raw data
from county_level.raw.chrr_health.load import load_chrr_health  # more generally: from county_level.raw.<data_source>.load import load_<data_source>
raw = load_chrr_health(data_dir = raw_dir)
raw.head()
```

```python
# loading cleaned data
from county_level.processed.chrr_health.clean import clean_chrr_health  # more generally: from county_level.processed.<data_source>.clean import clean_<data_source>
cleaned = clean_chrr_health(data_dir = raw_dir, out_dir = processed_dir)
cleaned.head()
```
